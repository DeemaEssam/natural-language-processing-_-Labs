Write a function that accepts a genre url and returns the genre name based on your experimentation above






my note:

url.strip().split('|'): This removes any leading/trailing spaces and splits the string into a list of individual genre URLs.

genre.split('/')[-1]: For each genre URL, this extracts the last part of the URL after the last '/', which is the genre name (e.g., 'fiction', 'mystery').

"|".join(...): The genre names are then joined back together with '|' as the separator.


````python
def split_and_join_genres(url):
    # Split the genre string by '|' to get individual genres
    genres = url.strip().split('|')
    # Extract the genre name from each genre URL and join them back with '|'
    return "|".join([genre.split('/')[-1] for genre in genres])

# Example usage
test_genre_string = df.genre_urls[0]
result = split_and_join_genres(test_genre_string)
print(result)

````

<div class="exercise"><b>Exercise</b></div>

Lets get the best-rated book(s) for every year in our dataframe.

````python
#Using .groupby, we can divide the dataframe into subsets by the values of 'year'.
#We can then iterate over these subsets
# your code here

# Group the DataFrame by 'year'
grouped_by_year = df.groupby('year')

# Create a list to store the best-rated books for each year
best_books_list = []

# Iterate over the grouped data and find the book(s) with the maximum rating for each year
for year, group in grouped_by_year:
    # Get the book(s) with the maximum rating for the current year
    max_rating = group['rating'].max()
    best_books = group[group['rating'] == max_rating]

    # Append the best book(s) to the list
    best_books_list.append(best_books)

# Concatenate the list into a final DataFrame
best_books_per_year = pd.concat(best_books_list).reset_index(drop=True)

# Display the result
print(best_books_per_year)

````
